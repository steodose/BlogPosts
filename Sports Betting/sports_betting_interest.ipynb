```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import altair as alt
```

```python
# Load data
def load_data():
        df = pd.read_csv('/Users/Stephan/Desktop/Python/Odds-Converter/google_trends.csv')
        return df

df = load_data()

df
```

```python
# Limit data to only 2015 and newer

# Convert the 'month' column to datetime
df['month'] = pd.to_datetime(df['month'])

# Filter for rows where the year is 2015 or later
df_filtered = df[df['month'].dt.year >= 2015]

# Display the first few rows of the filtered dataframe
df_filtered.head()
```

## Bar Chart

Using the Altair library

```python

# Convert the 'month' column back to string format 'YYYY-MM' for display purposes
df_filtered['month_str'] = df_filtered['month'].dt.strftime('%b-%Y')

# Create a condition for coloring the bars
# condition = alt.condition(
#     alt.datum.month >= pd.to_datetime('2018-05-01'),  # Condition for the date
#     alt.value('steelblue'),  # Color for True (May 2018 and beyond)
#     alt.value('grey')   # Color for False (Before May 2018)
# )

# Create a new column for conditional color logic
df_filtered['is_after_may_2018'] = df_filtered['month'] >= '2018-05-01'


# Define your title and subtitle
title = "Sports Betting Interest is on the Rise"
subtitle = "Google searches for sports betting have increased since a May 2018 Supreme Court decision cleared the way for legalization"
footer = "Source: Google Trends | Graphic: @steodosescu"

# Combine title, subtitle, and footer
combined_title = alt.TitleParams(
    text=title, 
    subtitle=subtitle,
    anchor='middle',  # Center the title
    color='black',    # Title color
    fontSize=20,      # Title font size
    subtitleColor='gray',  # Subtitle color
    subtitleFontSize=15,   # Subtitle font size
)



# Create bar chart with conditional coloring and Title fields
bar = alt.Chart(df_filtered).mark_bar().encode(
    x = alt.X('month_str:O', sort=alt.EncodingSortField(field='month', order='ascending')),
    y = 'search_interest:Q',
    color=alt.condition(alt.datum.is_after_may_2018, 
                        alt.value('steelblue'), 
                        alt.value('grey'))
).properties(
    title=combined_title
)

# Create line chart
line = alt.Chart(df_filtered).mark_line(color='red').transform_window(
    # The field to average
    rolling_mean='mean(search_interest)',
    # The number of values before and after the current value to include.
    frame=[-12, 0]
).encode(
    x=alt.X('month_str:O', 
            sort=alt.EncodingSortField(field='month', order='ascending'),
            axis=alt.Axis(title='')),
    y=alt.Y('rolling_mean:Q',
             axis=alt.Axis(title='Relative Interest over Time', titleFontWeight='bold'))
)

# Combine bar and line charts. Use scaleFactor = 2 to increase image size and resolution
combined_chart = (bar + line).properties(width=900).display(
    scaleFactor=2
)

# Display the chart
combined_chart.display()

```
